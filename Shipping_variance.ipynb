```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
orders['purchase_date'] = orders['order_purchase_timestamp\t'].dt.date
orders['purchase_time'] = orders['order_purchase_timestamp\t'].dt.time
```

```python
orders
```

```python
orders.drop(['order_purchase_timestamp\t'],inplace = True , axis = 1)
```

```python
orders_loop = ['order_approved_at',
              'order_delivered_carrier_date',
              'order_delivered_customer_date',
              'order_estimated_delivery_date']

for col in orders_loop:
    orders[col]=pd.to_datetime(orders[col])
    orders[col + ' date'] = orders[col].dt.date
    orders[col + ' time'] = orders[col].dt.time
    
```

```python
orders.columns.tolist()
```

```python
orders.drop('order_purchase_timestamp', axis=1, inplace=True)

orders.drop('order_approved_atdate', axis=1, inplace=True)

orders.drop('order_approved_attime', axis=1, inplace=True)

orders.drop('order_delivered_carrier_datedate', axis=1, inplace=True)

orders.drop('order_delivered_carrier_datetime', axis=1, inplace=True)

orders.drop('order_delivered_customer_datedate', axis=1, inplace=True)

orders.drop('order_delivered_customer_datetime', axis=1, inplace=True)

orders.drop('order_estimated_delivery_datedate', axis=1, inplace=True)

orders.drop('order_estimated_delivery_datetime', axis=1, inplace=True)
```
